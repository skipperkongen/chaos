# The lorenz 

```python
import numpy as np
from matplotlib import pyplot as plt

def lorenz(x, y, z, s=10, r=28, b=2.667):
    """
    Given:
       x, y, z: a point of interest in three dimensional space
       s, r, b: parameters defining the lorenz attractor
    Returns:
       x_dot, y_dot, z_dot: values of the lorenz attractor's partial
           derivatives at the point x, y, z
    """
    x_dot = s*(y - x)
    y_dot = r*x - y - x*z
    z_dot = x*y - b*z
    return x_dot, y_dot, z_dot

def simulate(x, y, z, s=10, r=28, b=2.667, n_steps=10000, dt=0.01):
    xs, ys, zs = [], [], []
    for _ in range(n_steps):
        xs.append(x)
        ys.append(y)
        zs.append(z)
        x_dot, y_dot, z_dot = lorenz(x, y, z, s, r, b)
        x += dt*x_dot 
        y += dt*y_dot 
        z += dt*z_dot         
    return np.array(list(zip(xs, ys, zs)))

def plot(x0, y0, s=10, r=28, b=2.667, n_steps=10000):
    ts = np.arange(n_steps)
    xs, ys, zs = simulate(x0, y0, z0, s=s, r=r, b=b, n_steps=n_steps)
    
    plt.plot(ts, xs, color='blue')
    plt.plot(ts, ys, color='red')
    plt.plot(ts, zs, color='green')
    plt.show()
    

n_steps = 1200
out1 = simulate(1.001, 3.001, 4.001, n_steps=n_steps)
out2 = simulate(1.002, 3.002, 4.002, n_steps=n_steps)
plt.title('Distances per dimension')
plt.plot(out1 - out2)
plt.show()

plt.title('Euclidean distance')
plt.plot(np.linalg.norm(out1-out2, axis=1))
plt.show()
```
